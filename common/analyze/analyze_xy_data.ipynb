```python
import matplotlib.pyplot as plt
import csv


def analyze_csv(csv_file):
    # Data storage
    local_x, local_y = [], []

    # Read the CSV file
    with open(csv_file, newline='') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the header
        for row in reader:
            local_x.append(float(row[0]))  # Assuming X values are in the first column
            local_y.append(float(row[1]))  # Assuming Y values are in the second column

    # Plot data
    plt.figure(figsize=(8, 5))
    plt.plot(local_x, local_y, label="Trajectory (XY)", marker='o', linestyle='-')

    # Labels and legend
    plt.xlabel("X Position (m)")
    plt.ylabel("Y Position (m)")
    plt.title("XY Trajectory")
    plt.legend()
    plt.grid(True)
    plt.axis("equal")  # Ensure equal scaling for correct trajectory representation
    plt.show()
```

```python
analyze_csv("coords_data/gps_data_300.csv")

```
